### Data creation
We select audio recordings from the Google Speech Command (GSP) commands as sound source, and use Pyroomacoustics as the platform to generate the received signal in a 7\*7\*3 room, considering the room reveberation and noise (snr=30db). 

The sound source is placed in the distance randomly selected between 0.5m-2m. The genenrated audio signal is in 16000Hz.

The angle of the sound source is from 0-359 degree with 1 degree space. For training, we generate 100 samples for each degree, and 25 samples are  generated for testing.


```python
from dataset.data_creation import data_generation
dataset = pra.datasets.GoogleSpeechCommands(basedir='./data/google_speech_commands',download=True  # Set to True if have not downloaded
selected_word = ['one','two','three','four','five','six','seven','eight','nine','zero']
matches = dataset.filter(word=selected_word)
print(matches)
print("Number of '%s' samples : %d" % (selected_word, len(matches)))
for num_source in range(1,2):
    print(f"Generating source {num_source}")
    data_generation(num_source=num_source,dataset=dataset,train=True,cohenet=True, save_dir='./GSP_data/train/coherent/')
```
